```python
from numpy import array
from keras.preprocessing.text import Tokenizer
from nltk.tokenize import word_tokenize
from tensorflow.keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Embedding
from keras.layers import LSTM
```

```python
data = """Football is an International sports which was born on England\n
I like so many footballers ever played and watch them on TV\n
Many football clubs are played with different styles\n
Very few clubs are got success and praised by people\n
Manchester Utd is an old club and is favorite to me\n"""
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
X = tokenizer.texts_to_sequences([data])[0]
X
```

```python
vocab_size = len(tokenizer.word_index) + 1
print("vocabulary size:",vocab_size)
vocab_size
```

```python
vocab_size = tokenizer.word_index.items()
print("vocabulary:",vocab_size)
len(vocab_size)
```

```python
sequences = []

for i in range(1, len(X)):
    sequence = X[i-1:i+1]
    sequences.append(sequence)
print('Total Sequence:', len(sequences))
```

```python
sequences = array(sequences)
X, y = sequences[:,0], sequences[:,1]
print("X",X)
print('\t')
print("y",y)
```

```python
y = to_categorical(y, vocab_size)
y[:6]
```

```python
model = Sequential()
model.add(Embedding(vocab_size, 10, input_length= 1))
model.add(LSTM(30))
model.add(Dense(vocab_size, activation = 'softmax'))

model.summary()
```

```python
import tensorflow as tf
```

```python
model.compile(loss= tf.keras.losses.categorical_crossentropy, optimizer = tf.keras.optimizers.Adam(), metrics =['accuracy'])
```

```python
model.fit(X, y, epochs= 100)
```

```python
def generate_seq(model, tokenizer, enter_text, n_pred):
    in_text, result = enter_text, enter_text
    
    for _ in range(n_pred):
        encoded = tokenizer.texts_to_sequences([in_text])[0]
        y_hat = model.predict(encoded)
        outword = ''
    
        
        for index, word in tokenizer.word_index.items():

          if word == y_hat:
            out_word = index
            break
            
        in_text, result = outword, result + ' '+ out_word


    return result
```

```python
print(generate_seq(model, tokenizer, 'football', 5))
```
